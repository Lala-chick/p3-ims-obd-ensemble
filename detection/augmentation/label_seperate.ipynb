```python
import json 
import copy
```

```python
def seperate_json(input_file_path, output_file_path, selected_labels = [0, 1, 2, 3, 4, 5, 6, 6, 7, 8, 9, 10]):
    with open(input_file_path, 'r') as f:
        ori_data = json.load(f)
    new_json =  copy.deepcopy(ori_data)
    new_json["annotations"] = []
    for ann in ori_data["annotations"]:
        if ann["category_id"] in selected_labels:
            new_json["annotations"].append(ann)

    with open(output_file_path, 'w') as f:
        json.dump(new_json, f, indent="\t")
```

```python
seperate_json('./train_all.json','./train_label_0.json', [0])
```
